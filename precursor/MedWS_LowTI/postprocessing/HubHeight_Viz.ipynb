## 7x2 case

```python
amrdir = '/lustre/orion/cfd162/scratch/lcheung/ALCC_Frontier_WindFarm/MedWS_LowTI/precursor6_7kmX2km_5m/post_processing/'
ncfile = 'samplePlane_100000.nc'  # netcdf filename
iters  = [0]         # timesteps to extract
varnames = ['velocityx', 'velocityy']  # variables to extract
db     = ppsamplexr.getPlaneXR(amrdir+ncfile, iters, varnames, verbose=1)
```

```python
plt.figure(figsize=(12,3), dpi=125)
clevels=np.linspace(7, 11,121)
iplane=1
Uh=np.sqrt(db['velocityx'][0][iplane,:,:]**2+db['velocityy'][0][iplane,:,:]**2)
plt.contourf(db['x'][iplane,:,:], db['y'][iplane,:,:], Uh, levels=clevels, cmap='coolwarm')
plt.colorbar()
plt.xlabel('x [m]')
plt.ylabel('y [m]')
plt.title('Horizontal velocity')
plt.axis('scaled')
```
